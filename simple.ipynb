```python
import math
```

```python
# 0以上整数x「未満」の素数をリストに格納して返す
def primes(x):
    if x < 2: return []

    primes = [i for i in range(x)]
    primes[1] = 0 # 1は素数ではない

    # エラトステネスのふるい
    for prime in primes:
        if prime > math.sqrt(x): break
        if prime == 0: continue
        for non_prime in range(2 * prime, x, prime): primes[non_prime] = 0

    return [prime for prime in primes if prime != 0]
```

```python
num=100000
sosu = primes(num)
#print(sosu)
```

```python
array = []
for x in range(1, num):
    array.append(x)
#print(array)
```

```python
x_array=[]
y_array=[]
not_x_array=[]
not_y_array=[]
radius = 0.0
for ang in range(0, num):
    if ang in sosu:
        radius += 10
        rad = math.radians(ang)
        x = radius * math.cos(rad)
        y = radius * math.sin(rad)
        x_array.append(x)
        y_array.append(y)
    else:
        radius += 10
        rad = math.radians(ang)
        x = radius * math.cos(rad)
        y = radius * math.sin(rad)
        not_x_array.append(x)
        not_y_array.append(y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.plot(x_array, y_array, color='k', marker='.', ls="", markersize=0.5)
plt.savefig('./prime.png')
plt.figure(figsize=(8,8))
plt.plot(not_x_array, not_y_array, color='k', marker='.', ls="", markersize=0.5)
plt.savefig('./not_prime.png')
```
